```python
from matplotlib import pyplot as plt  #Importing pylot as plt from the matlplotlib module
input_samples = [(2, 3), (3, 3), (3, 4), (1, 4), (4, 1), (4, 3)] #Input Data Sample Set 1
input_samples1 = [(0, 0), (0, 3), (1, 1), (1, 2), (2, 1), (2, 2)] #Input Data Sample Set 2
```

```python
def simpleplot(input_samples, input_samples1):
    x_points,y_points = zip(*input_samples)  #zip function is seperating the x_points and y_points for the each sample and giving us seperate lists.
    plt.plot(x_points, y_points, 'ks', label = "Input Data Sample Set 1") #ks refers to black squares

    x_points1,y_points1 = zip(*input_samples1)  #zip function is seperating the x_points and y_points for the each sample and giving us seperate lists.
    plt.plot(x_points1, y_points1, 'co', label = "Input Data Sample Set 2") #co refers to cyan circles

    plt.xlabel("X - Axis")
    plt.ylabel("Y - Axis")
    plt.axis([-1,5, -1,5]) #As suggested to select axis between [-1,5].
    plt.title("Input data samples plot")
    plt.legend()
    plt.show()

simpleplot(input_samples, input_samples1) #Calling the function initially to show the data samples on a plot.
```

```python
def classifier_data(thx,thy,input_samples, input_samples1):
    correctly_classified = 0
    total_samples = len(input_samples)+len(input_samples1)
    c1=[]
    c2=[]
    for i in input_samples: #first sample set is considered and each point is classified into either of the classes C1 and C2 based on the given condition
        if i[0] >= thx and i[1]>=thy:
            c1.append(i)
        else:
            c2.append(i)

    for i in input_samples1: #Second sample set is considered and each point is classified into either of the classes C1 and C2 based on the given condition
        if i[0] >= thx and i[1]>=thy:
            c1.append(i)
        else:
            c2.append(i)

    for i in input_samples: #counting the correctly classified samples with the first set.
        if i in c1:
            correctly_classified+=1
    for j in input_samples1: #counting the correctly classified samples with the second set.
        if j in c2:
            correctly_classified+=1

    classification_accuracy = classification_accuracy_calculator(correctly_classified, total_samples) #function to find the classification accuracy.
    print("The Classification Accuracy percentage when the thresholds thx = {} and thy = {} rounded to 2 decimals is {}".format(thx,thy,classification_accuracy))

    plot_with_thresholds(thx, thy, input_samples, input_samples1) #calling the function to plot the graph with Input data samples, thresholds thx and thy.
    print("*-"*40)
    return

def classification_accuracy_calculator(correctly_classified, total_samples):
    accuracy_in_percentage = round((correctly_classified/total_samples)*100,2) #rounding the answer to 2 decimals. Classification Accuracy formula.
    return accuracy_in_percentage

def plot_with_thresholds(thx,thy, input_samples, input_samples1):
    x_points,y_points = zip(*input_samples)  #zip function is seperating the x_points and y_points for the each sample and giving us seperate lists.
    plt.plot(x_points, y_points, 'gs', label = "Input Data Sample Set 1") #gs refers to green squares

    x_points1,y_points1 = zip(*input_samples1)  #zip function is seperating the x_points and y_points for the each sample and giving us seperate lists.
    plt.plot(x_points1, y_points1, 'ro', label = "Input Data Sample Set 2") #ro refers to red circles

    plt.axvline(x=thx, color = "blue",  label = "thx = {}".format(thx)) #Referred documentation to find how to plot the verticle line
    plt.axhline(y=thy, color = "brown", label = "thy = {}".format(thy)) #Referred documentation to find how to plot the horizontal line

    plt.xlabel("X - Axis")
    plt.ylabel("Y - Axis")
    plt.axis([-1,5, -1,5]) #As suggested to select axis between [-1,5].
    plt.title("Input Data samples plot with Thresholds thx and thy")
    plt.legend()
    plt.show()
```

```python
def user_inputs(): #this will take the user inputs, start the classification, performs the accuracy calulation and plotting.
    while True:
        try:
            thx = input("Enter the X - Axis threshold Value OR x or X to exit: ") #threshold thx value
            thy = input("Enter the Y - Axis threshold Value OR x or X to exit: ") #threshold thy value
            if thx in ['x','X'] or thy in ['x','X']:
                print("You have entered the option to exit. Thank you for testing my Code") #if user inputs x or X it should stop execution.
                break
            else:
                thx = float(thx) #inputs can be decimals.
                thy = float(thy)
                classifier_data(thx,thy, input_samples, input_samples1) #calling the function to find the classification accuracy.
        except:
            print("Error!!!! You have entered characters instead of numbers or x. Please try again by giving numbers as inputs")
```

```python
user_inputs() #Calling the function user_inputs to start working on the program.
```
